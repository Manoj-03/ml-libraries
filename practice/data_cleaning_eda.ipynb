Correlation Matrix

What It Shows
- A **correlation matrix** displays the **Pearson correlation coefficients** between different features in a dataset (such as **ID, Age, Weight, Year**).
- The values range from **-1 to 1**.
- It shows how strongly two features are related to each other.

Interpretation of Values
- **1 or -1**  
  - Perfect correlation  
  - **1**: Perfect positive correlation  
  - **-1**: Perfect negative correlation  

- **0**  
  - No correlation  
  - The features are not related.

- **Positive values**  
  - As one feature increases, the other feature **also tends to increase**.

- **Negative values**  
  - As one feature increases, the other feature **tends to decrease**.

Summary: 
A correlation matrix helps identify **relationships between features** and shows whether they move together, move in opposite directions, or have no relationship at all.

```python
data.groupby(['City', 'Medal']).size().unstack() # size() counts number of records in each City–Medal combination, unstack() - it is used to reshape the data into a dataframe (readable format)
```

Sort the medal counts per city based on the total no.of medals(Bronze, Silver, Gold)

```python
medal_count = data.groupby(['City', 'Medal']).size().unstack()
```

```python
medal_count['Total'] = medal_count[['Bronze','Gold','Silver']].sum(axis=1)
```

```python
medal_count.sort_values(by='Total', ascending=False)
```
