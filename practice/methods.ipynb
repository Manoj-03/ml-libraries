```python
import pandas as pd
```

```python
dict1 ={
    'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv','Mittal','Archana'],
    'Marks':[98,89,99,87,90,83,99],
    'Gender':['Male','Female','Female','Male','Male','Female','Female']
}
df1=pd.DataFrame(dict1)
df1
```

```python
# 1. displaying top 3 rows of dataset
df1.head(3)
```

```python
# 2. check last 3 rows of dataset
df1.tail(3)
```

```python
# 3. Find shape of dataset(Number of rows and cols)
print(df1.shape)
print('Number of rows', df1.shape[0])
print('Number of cols', df1.shape[1])
```

```python
# 4. get info about dataset like no.of rows, no.of cols, datatype of each col and memory
df1.info()
```

```python
# 5. check null of dataset
# df1.isnull() # prints boolean
df1.isnull().sum() # prints as summary col name and total count
# df1.isnull().sum(axis=1) # prints null values row wise
# df1.isnull().sum(axis=0) # prints null values col wise
```

```python
# 6. get overall statistics of data frame
# df1.describe() # by default describe() method gives only statistics of numerical value col
# to print inlcuding numerical and categorical use include='all'
df1.describe(include='all')
```

```python
# 7. find unique values of gender col
df1['Gender'].unique()
```

```python
# 8. find number of unique values of gender col
df1['Gender'].nunique()
```

```python
# 9. display count of Unique values in gender col
df1['Gender'].value_counts()
```

```python
# 10. find total no.of students having marks between 90 and 100
# df1[df1['Marks']>=90] # shows only records who scored 90 or more marks
df1[(df1['Marks']>=90) & (df1['Marks']<=100)]

# using between method
# df1['Marks'].between(90,100) # both 90 and 100 are included
```

```python
# 11. find average of marks
df1['Marks'].mean()

# df1['Marks'].min() # minimum of marks
# df1['Marks'].max() # maximum of marks
```

```python
# 12. Apply method
def marks(x):
    return x//2

df1['Marks'].apply(marks)

df1['Half_Marks'] = df1['Marks'].apply(marks)
df1
```

```python
df1['Marks'].apply(lambda x: x/2)

# df1['Name'].apply(len) # prints length of string
```

```python
# 13. Map Function
# need to change Male - 1 and Female - 0

df1['Male_Female'] = df1['Gender'].map({'Male': 1, 'Female': 0})
df1
```

```python
# 14. drop the cols
# df1.drop('Male_Female',axis =1) # it will drop single column
df1.drop(['Male_Female','Half_Marks'],axis=1,inplace=True) # to drop multiple cols use list
df1
```

```python
# 15. print names of columns
df1.columns

# df1.index # prints the index's start, stop and step
```

```python
# 16. sort the data frame as per marks column
# df1.sort_values(by='Marks', ascending=False) # by default ascending order
df1.sort_values(by=['Marks','Gender'], ascending=False) # sorting based on multiple values
```

```python
# 17. Display name and marks of female students
df1[df1['Gender'] == 'Female'][['Name','Marks']]
```
