```python
import pandas as pd
import numpy as np

df = pd.read_csv('../sample_data/455_461.csv')
df.head()
```

```python
# checking Data types
df.dtypes
```

```python
# check missing values
df.isna().sum()
```

```python
# Drop unwanted index column
df.drop(columns=["Unnamed: 0"], inplace=True)
df.head()
```

```python
# Convert Timestamp to Datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'],errors='coerce')
df.head()
```

```python
# Convert Date of Birth to Datetime
df['Date of birth'] = pd.to_datetime(df['Date of birth'],errors='coerce')
df.head()
```

```python
# Graduation Percentage(cgpa to %)
# first inspect values
df["Graduation Percentage (convert cgpa to percentage)"].unique()
```

```python
# function
def clean_percentage(val):
    if pd.isna(val):
        return np.nan
    val = str(val).replace('%',"").strip()
    try:
        val = float(val)
        if val <= 10:        # CGPA
            return val * 9.5
        return val
    except:
        return np.nan
```

```python
# creating a new column
df["Graduation Percentage"] = df["Graduation Percentage (convert cgpa to percentage)"].apply(clean_percentage)
```

```python
# drop old col
df.drop(columns=["Graduation Percentage (convert cgpa to percentage)"], inplace=True)
```

```python
df.head()
```

```python
# Clean Experience col
# inspect values
df["Experience (in years, freshers/pursuing students fill 0)"].unique()
```

```python
df["Experience (in years)"] = (
    df["Experience (in years, freshers/pursuing students fill 0)"]
    .astype(str)
    .str.extract(r"(\d+\.?\d*)") # Extracts integer or decimal numbers from mixed text (e.g., "3 years", "0.4 yrs") using regex
    .astype(float)
    .fillna(0)
)
```

```python
df.drop(columns=["Experience (in years, freshers/pursuing students fill 0)"], inplace=True)
```

```python
df.head()
```

```python
# Clean Distance column
df["Distance you travel everyday to reach Innomatics (in km only)"].unique()
```

```python
df["Daily Travel Distance (km)"] = (
    df["Distance you travel everyday to reach Innomatics (in km only)"]
    .str.lower()
    .str.replace("km", "", regex=False)
    .str.strip()
)

df["Daily Travel Distance (km)"] = pd.to_numeric(
    df["Daily Travel Distance (km)"], errors="coerce"
)

df["Daily Travel Distance (km)"].fillna(
    df["Daily Travel Distance (km)"].median(),
    inplace=True
)
```

```python
df.drop(columns=["Distance you travel everyday to reach Innomatics (in km only)"], inplace=True)
```

```python
df.head()
```

```python
# Standardize Gender Column
df["Gender"] = df["Gender"].str.capitalize()
```

```python
# Handle Missing Industry Values
df["Industry"] = df["Industry"].fillna("Not Specified")
```

```python
# to save the cleaned data
df.to_csv('../sample_data/cleaned_dataset.csv', index=False)
```
