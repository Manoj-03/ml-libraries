Data Visualization:
 - Converting Raw Data to Visuals using charts,graphs..
 - It helps data to clear,simple and understand

Matplotlib:
- It is a python library used for data visualization
- By using this we can create Line Charts, Bar Graphs, Scatter Plots, Histograms, Pie-Charts, 3D Plots
- We can fully customize our plots like colors, styles, title, labels and legends
- Two ways to use:
    1. Pyplot API: Simpler and commonly used method.
    2. Object-Oriented API: Better used for complex and customized visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y) # it will plot (1,4), (2,5), (3,6)
plt.grid() # to show grid lines
plt.show()
```

Univariate:
- Focuses on one variable at a time

Univariate - Numerical

```python
data = {
    "Salary" : [25000,35000,39000,42000,45000,50000,55000,52000,17000,20000,31000,65000,70000,10000,75000,47000,27000,63000,57000,23000]
}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
# Line Plot
plt.plot(df['Salary'],color='red',marker='o',linestyle='--',linewidth='2')
plt.grid()
plt.show()
''' 
    Params:
        color='red' → line color
        marker='o' → marks each data point
        linestyle='--' → dashed line
        linewidth=2 → line thickness
        plt.grid() → adds grid for better visual clarity
'''
```

```python
# Histogram - frequency of our data distribution 
plt.hist(df['Salary'],bins=5,color='green') # Divides Salary values into 5 bins (ranges)
plt.show()
```

```python
# Box plot
plt.boxplot(df['Salary'])
plt.show()
```

Univariate - Categorical

```python
df['dept'] = ['HR', 'IT', 'Finance','HR','Finance', 'IT', 'Finance', 'IT', 'HR','IT'] * 2
```

```python
df.head()
```

```python
df.info()
```

```python
# Pie-Chart
count = df['dept'].value_counts()
count
```

```python
plt.pie(count,labels=count.index, autopct='%1.1F',explode=[0,0.1,0.2])
plt.axis('equal')
plt.show()

# label shows category names for each slice
# Each slice represents a category and its relative frequency
# autopct displays percentage values on each slice
# explode separates selected slices for emphasis
# axis('equal') ensures the pie chart is drawn as a perfect circle
```

```python
# count plot
plt.bar(count.index,count,color=['green','red','black'])
plt.show()
```

Bivariate:
- Focuses on relationship between two columns

Bivariate: numerical - numerical

```python
df['Age'] = [22, 58, 19, 35, 27, 63, 24, 41, 18, 29,55, 26, 47, 31, 20, 38, 60, 23, 44, 28]
```

```python
df.head()
```

```python
# scatter plot - analyses the direct relation like salary increases with age..
plt.scatter(df['Age'],df['Salary'],color='red')
```

```python
sort_age = df.sort_values('Age')
```

```python
# Line plot - before plotting sort the values
plt.plot(sort_age['Age'],df['Salary'],color='red',marker='o',linewidth='2')
plt.grid()
plt.show()
```

```python
# Bar chart
plt.bar(sort_age['Age'],df['Salary'], align='center')
```

Bivariate: Numerical - Categorical

```python
hr_sal = df[df['dept'] == 'HR']['Salary']
hr_sal
```

```python
it_sal = df[df['dept'] == 'IT']['Salary']
it_sal
```

```python
fin_sal = df[df['dept'] == 'Finance']['Salary']
fin_sal
```

```python
# Box plot
plt.boxplot([hr_sal,it_sal,fin_sal],labels=['HR','IT','Finance'])
plt.show()
```

```python
# pie chart
sal_by_dept = df.groupby('dept')['Salary'].sum()
sal_by_dept
```

```python
plt.pie(sal_by_dept, labels=sal_by_dept.index,autopct='%1.2f') # other param - shadow(true/false),explode
plt.show()
```

```python
# bar plot - For avg salary of each dept
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)
```

```python
plt.bar(['HR','IT','Finance'],[hr_mean,it_mean,fin_mean])
plt.show()
```

Multivariate:
- Relationship between 3 columns

Multivariate: Numerical - Numerical - Numerical

```python
df['experience'] = [1.6, 1, 2, 3, 5, 7, 4, 6, 8, 10,1.5, 2.5, 4.5, 9, 12, 15, 0.5, 11, 18, 20]
```

```python
df.head()
```

```python
# Bubble Plot - a scatter plot, where third numerical col will define the bubble size
plt.scatter(df['Age'],df['Salary'],s=df['experience']*20,color='skyblue',edgecolors='black')
plt.title('Age vs Salary vs Experience')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

Multivariate: 2 numerical and 1 categorical

```python
plt.scatter(df['Age'],df['Salary'], c=df['dept'].map({'HR':'yellow','IT':'blue','Finance':'orange'}))
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary vs Experience')
plt.show()
```

```python
# for legend - 2 numerical and 1 categorical

color = {'HR':'yellow','IT':'blue','Finance':'orange'}
for dept,color in color.items():
    df_dept = df[df['dept'] == dept]
    plt.scatter(df_dept['Age'],df_dept['Salary'], c = color, label = dept)
plt.legend()
```

Object Oriented API
- allows sub or multiple plots in a figure

```python
figure,axis = plt.subplots(1,3,figsize=(10,5)) # fig,axs are objects
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,5))

# Line Plot
axs[0].plot(sort_age['Age'],df['Salary'], color='red',marker='*',linewidth='2',markersize=2)
axs[0].grid()
axs[0].set_title('Line Plot')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Salary')

# Histogram
axs[1].hist(df['Salary'],bins=5,color='skyblue')
axs[1].set_title('Histogram')
axs[1].set_xlabel('Salary')
axs[1].set_ylabel('Frequency')

# Box Plot
axs[2].boxplot(df['Salary'])
axs[2].set_title('Histogram')
axs[2].set_xlabel('Salary')

plt.savefig('multipleplots.png') # to save image - output
```

```python
# visualisation for all numerical columns

data2 = {
    'Year': [2020, 2021,2022, 2023],
    'Sales': [100, 150, 200, 250],
    'Profit': [20, 30, 40, 50],
    'Expenses': [80, 120, 160, 200]
}
```

```python
df2 = pd.DataFrame(data2)
```

```python
df2
```

```python
plt.plot(df2['Year'], df2['Sales'],label = 'Sales')
plt.plot(df2['Year'], df2['Profit'],label = 'Profit')
plt.plot(df2['Year'], df2['Expenses'],label = 'Expenses')

plt.title('Analysis')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

3D PLOT

```python
# 3-D Plot
ax = plt.axes(projection='3d')
ax.scatter(df['Age'],df['Salary'],df['experience'])
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_zlabel('Experience')
plt.show()
```
