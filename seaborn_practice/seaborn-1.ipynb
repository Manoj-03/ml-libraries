Seaborn allows us various ways of styling themes are as follows:
- white
- dark
- whitegrid
- darkgrid
- ticks

set_title():
- used to set the title of an individual axes
- Params: 
    - label: The title string, 
    - fontsize: Font size of the title
    - color: Color of the title
    - loc (left, center, right): Customize appearance 

set_context():
- Control the scale of plot elements in Seaborn (like labels, lines, and markers)
- It changes the "context" of the plot
- It affects the whole plot style
- Params:
    - context: Can be 'paper', 'notebook', 'talk', 'poster'
        - 'paper' → smaller labels, suitable for publications
        - 'notebook' → default, good for Jupyter notebooks
        - 'talk' → larger labels for presentations
        - 'poster' → very large labels for big posters
    - font_scale: Multiplier for font size. Default is 1.0
    - rc: Optional dictionary to override specific matplotlib parameters

```python
sns.set_style('white')
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex')
sns.despine() # removes the top and right spines(lines)
```

```python
sns.set_style('dark')
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex')
sns.despine() # removes the top and right spines(lines)
```

```python
sns.set_style('white')
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex')
sns.despine(left = True) # removes the top and right, left spines(lines)
```

```python
sns.set_style('whitegrid')
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex')
# sns.despine() # removes the top and right spines(lines)
```

```python
sns.set_style('ticks')
sns.set_context('notebook') 
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex')
sns.despine()
```

```python
sns.set_style('ticks')
sns.set_context('paper')
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex')
sns.despine()
```

```python
sns.set_style('ticks')
sns.set_context('talk')
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex')
sns.despine()
```

```python
sns.set_style('ticks')
sns.set_context('poster')
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'sex')
sns.despine()
```
