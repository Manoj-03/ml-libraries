```python
# loading a dataset
import pandas as pd
df = pd.read_csv('../sample_data/data_cleaning_dataset.csv')
print(df)
```

```python
# cleaning empty cells
# 1. remove rows with empty cells - uses dropna() method, it returns a new DataFrame, and will not change the original.
df_dropped = df.dropna()
print("\nDataFrame after dropping rows with empty cells:")
print(df_dropped.to_string())

# Note: if you want to change the original DataFrame, use the inplace parameter:
# df.dropna(inplace=True)
```

```python
# 2. fill empty cells with a specific value - uses fillna() method
# df_filled = df.fillna(value="No Value")

print("\nDataFrame after filling empty cells:")
df_filled = df.fillna(value=130)
print(df_filled.to_string())
```

```python
# 3. Replace only certain empty cells - using a dictionary to fill different values in different columns
# df_filled_specific = df.fillna(value={"Calories": 0, "Duration": 100})

print("\nDataFrame after filling empty cells with specific values:")
df_filled_specific = df.fillna(value={"Calories": 100}) # only filling empty cells in 'Calories' column with 100
print(df_filled_specific.to_string())
```

```python
# 4. Replace empty cells with the mean, median, or mode of the column
import pandas as pd
df = pd.read_csv('../sample_data/data_cleaning_dataset.csv')
mean_calories = df['Calories'].mean()
print("Mean calories:", mean_calories)
df.fillna(value={'Calories': mean_calories}, inplace=True) # filling empty cells in 'Calories' column with mean value
print("\nDataFrame after filling empty cells in 'Calories' with mean value:")
print(df.to_string())

# using median 
# median_calories = df['Calories'].median()
# df.fillna(value={'Calories': median_calories}, inplace=True)

# using mode
# mode_calories = df['Calories'].mode()[0]
# df.fillna(value={'Calories': mode_calories}, inplace=True)
```

```python
# Pandas - Cleaning Data of Wrong Format (in our case, date format i.e rows 26 and 22)
# To fix it, you have two options: remove the rows, or convert all cells in the columns into the same format.
import pandas as pd
df = pd.read_csv('../sample_data/data_cleaning_dataset.csv')

# convert the 'Date' column to datetime format - uses to_datetime() method
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
print("\nDataFrame after converting 'Date' column to datetime format:")
print(df.to_string())
```

```python
# removing rows with wrong format
# The result from the converting in the example above gave us a NaT value, which can be handled as a NULL value, and we can remove the row by using the dropna() method.
df.dropna(subset=['Date'], inplace=True)
print("\nDataFrame after removing rows with wrong format:")
print(df.to_string())
```

```python
# Pandas - Fixing Wrong Data
# "Wrong data" does not have to be "empty cells" or "wrong format", it can just be wrong, like if someone registered "199" instead of "1.99".

# in our dataset, the 'Duration' column should be between 30 and 60 minutes. here row 7 has a wrong value i.e 450

# 1. replace wrong data with a specific value
import pandas as pd
df = pd.read_csv('../sample_data/data_cleaning_dataset.csv')

# for small datasets, we can manually replace the wrong data i.e df.loc[row_index, 'column_name'] = correct_value
# df.loc[7, 'Duration'] = 45  # replacing wrong value 450 with 45

# for larger datasets, we can use conditional replacement
for x in df.index:
    if df.loc[x, 'Duration'] > 60:  # assuming duration should not be more than 60
        df.loc[x, 'Duration'] = 60  # replacing wrong value with max valid value
print("\nDataFrame after fixing wrong data in 'Duration' column:")
print(df.to_string())
```

```python
# 2. removing rows with wrong data - uses drop() method.
import pandas as pd
df = pd.read_csv('../sample_data/data_cleaning_dataset.csv')

for x in df.index:
    if df.loc[x, 'Duration'] > 60:  # assuming duration should not be more than 60
        df.drop(x, inplace=True)  # removing the row with wrong data
print("\nDataFrame after removing rows with wrong data in 'Duration' column:")
print(df.to_string())
```

```python
# Pandas - Removing Duplicates
# Duplicates can mess up your data analysis. in our dataset, row 11 and row 12 are same.

# duplicated() - used to find duplicates, returns a boolean Series denoting duplicate rows. Returns True for each row that is a duplicate, otherwise False.

import pandas as pd
df = pd.read_csv('../sample_data/data_cleaning_dataset.csv')
duplicates = df.duplicated()
print("\nBoolean Series denoting duplicate rows:")
print(duplicates.to_string())

# drop_duplicates() - removes duplicate rows, keeps the first occurrence by default.
df.drop_duplicates(inplace=True)
print("\nDataFrame after removing duplicate rows:")
print(df.to_string())
```
