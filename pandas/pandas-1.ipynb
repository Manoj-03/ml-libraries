```python
# pandas - Data Analysis Library used for data manipulation, analysis and cleaning
import pandas as pd
print("Pandas version:", pd.__version__)
```

```python
# Creating a simple DataFrame 
# DataFrame is a 2D labeled data structure with columns of potentially different types like a spreadsheet or SQL table
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
print(df)
```

```python
# pandas series - 1D labeled array capable of holding any data type
# It is similar to a column in a table
# created using Series() function - pd.Series() parameters: data, index, dtype, name, copy 

# creating a simple Series using a list
import pandas as pd
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(series)

# Labels - used to access elements in a Series
print("Element at index 2:", series[2])  # Accessing element at index 2

# create labels - we create using index parameter, you can specify custom labels for the Series
labels = ['a', 'b', 'c', 'd', 'e']
custom_series = pd.Series(data, index=labels) # or pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])
print(custom_series)
# Accessing element using custom label
print("Element with label 'c':", custom_series['c'])

# Key/Value Object - Series can be created using a dictionary where keys become labels and values become data
data_dict = {'x': 100, 'y': 200, 'z': 300}
dict_series = pd.Series(data_dict)
print(dict_series)
# Accessing element using key
print("Element with key 'y':", dict_series['y'])
```

```python
# Series Properties and Methods
import pandas as pd
data = [5, 10, 15, 20, 25]
series = pd.Series(data)
print("Series:\n", series)
# Properties
print("Index:", series.index)          # Index of the Series
print("Values:", series.values)        # Values of the Series
print("Data Type:", series.dtype)      # Data type of the Series
print("Size:", series.size)            # Size of the Series
print("Shape:", series.shape)          # Shape of the Series
```

```python
# dataframe - 2D labeled data structure with columns of potentially different types like a spreadsheet or SQL table
# created using DataFrame() function - pd.DataFrame() parameters: data, index, columns, dtype, copy

# Creating a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
print(df)
```

```python
# Data Frame Properties and Methods
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
print("DataFrame Index:", df.index)          # Index of the DataFrame
print("DataFrame Columns:", df.columns)      # Columns of the DataFrame
print("DataFrame Values:\n", df.values)      # Values of the DataFrame
print("DataFrame Data Types:\n", df.dtypes)  # Data types of each column
print("DataFrame Shape:", df.shape)          # Shape of the DataFrame
print("DataFrame Size:", df.size)            # Size of the DataFrame
```

```python
# locate row - loc attribute is used to access a group of rows and columns by labels or a boolean array
# Accessing row with label 2
print("Row with label 2:\n", df.loc[2]) # returns a Series representing the row

# Accessing multiple rows with labels 1 and 3
print("Rows with labels 1 and 3:\n", df.loc[[1, 3]]) # returns a DataFrame with the selected rows
```

```python
# iloc - index based selection - iloc attribute is used to access a group of rows and columns by integer position(s)
# Accessing row at index 0
print("Accessing row at index 0:\n", df.iloc[0])
print("First column:\n", df.iloc[:, 0])  # first column
print("First two rows and first two columns:\n", df.iloc[0:2, 0:2])  # first two rows and first two columns
```

```python
# named index - using index parameter while creating DataFrame to set custom row labels
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
row_labels = ['a', 'b', 'c', 'd']
df_named_index = pd.DataFrame(data, index=row_labels)
print(df_named_index)

# Accessing row using custom label
print("Row with label 'c':\n", df_named_index.loc['c'])
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    "Name": ["John", "Anna", "Peter", "Linda"],
    "Age": [28, None, 35, 32],
    "City": ["New York", "Paris", "Chicago", "Paris"],
    "Salary": [50000, 60000, 70000, 65000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Boolean filtering - filter rows based on condition
print("\nPeople with Age > 30:")
print(df[df["Age"] > 30])

# GroupBy operation - group data based on a column and perform aggregate functions
print("\nAverage Salary by City:")
print(df.groupby("City")["Salary"].mean())

# Categorical encoding - converting categorical variables into a format that can be provided to ML algorithms
# Using get_dummies() to perform one-hot encoding on 'City' column
df_encoded = pd.get_dummies(df, columns=["City"])
print("\nEncoded DataFrame:")
print(df_encoded)
```

```python
# reading csv - pandas provides read_csv() function to read data from a CSV file into a DataFrame
# Example: df = pd.read_csv('file_path.csv')

# use csv from sample data folder
import pandas as pd
df_csv = pd.read_csv('../sample_data/data.csv')
print("Printing using to_string():\n",df_csv.to_string()) # to_string() to print the entire DataFrame without truncation
```

```python
# without to_string(), large DataFrames are truncated
print("Printing without to_string():\n",df_csv)

""" difference between using to_string() and not using it is that to_string() prints the entire DataFrame without truncation, 
while without it, large DataFrames are truncated for better readability in the console output. """
```

```python
# max_rows - pandas option is to set the maximum number of rows to display when printing a DataFrame

# check the number of maximum returned rows  - You can check your system's maximum rows with the pd.options.display.max_rows statement.
print("Maximum number of rows to display:", pd.options.display.max_rows) # default is 60

''' 
    In my system the number is 60, which means that if the DataFrame contains more than 60 rows, 
    the print(df) statement will return only the headers and the first and last 5 rows.
'''

# set max_rows - You can change the maximum number of rows to display by setting the pd.options.display.max_rows option to your desired value.
pd.options.display.max_rows = 10  # set maximum rows to 10
print("Updated maximum number of rows to display:", pd.options.display.max_rows)
```

```python
# pandas - analyzing data frames
# viewing the data - head() and tail() functions are used to view the first and last n rows of a DataFrame respectively
# head() - returns the first n rows of the DataFrame, default is 5

df_head = pd.read_csv('../sample_data/data.csv')
print("First 5 rows of the DataFrame using head():\n", df_head.head())  # default n=5
# specifying number of rows to return
print("First 3 rows of the DataFrame using head(3):\n", df_head.head(3))  # first 3 rows

# tail() - returns the last n rows of the DataFrame, default is 5
print("Last 5 rows of the DataFrame using tail():\n", df_head.tail())  # default n=5
# specifying number of rows to return
print("Last 2 rows of the DataFrame using tail(2):\n", df_head.tail(2))  # last 2 rows
```
